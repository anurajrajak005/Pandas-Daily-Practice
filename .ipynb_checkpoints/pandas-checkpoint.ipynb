```python
A= [[1,2,3],[4,5,6],[7,8,9]]
print(A)
ws=pd.Series(A)
print(ws)
```

```python
A= [[1,2,3],[4,5,6],[7,8,9]]
print(A)
ws=pd.Series(A)
print(ws[1][1])
print(ws)
```

```python
A= [[1,2,3],[4,5,6],[7,8,9]]
print(A)
ws=pd.Series(A)
print(ws)
print(ws[1][1])
print(ws[0][1])
```

```python
A= [[1,2,3],[4,5,6],[7,8,9]]
print(A)
ws=pd.Series(A)
print(ws)
print(ws[1:3][0:3])
print(ws[0][1])
```

```python
Name = ["Amit","Rohan","Govind","Honey"]
print(Name)
score = [89,56,78,65]
print(score)
ds=pd.Series(score,index=Name)
print(ds)
```
