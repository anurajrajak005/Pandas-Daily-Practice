```python
data = df.groupby(['Gender'],as_index=False)['Amount'].sum()
data
```

```python
ax = sns.barplot(x='Gender',y='Amount',data=data,palette=sns.color_palette("Set1"))
plt.show()
```

## from above graph we can see that most the buyers are female

```python
ax=  sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
Amount_Age = df.groupby('Age Group',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

Amount_Age
```

```python
sns.barplot(x='Age Group',y='Amount',data=Amount_Age,palette=sns.color_palette("Set2"))
total = sum(Amount_Age['Amount'])
#enumerate() return value & index
for i , v in enumerate(Amount_Age['Amount']): 
    per = v / total * 100
    plt.text(i,v,f'{per:.1f}%',ha='center',va='bottom')
plt.show()
```

## from above graph we can see that most the buyers age group between 26 to 35

```python
plt.figure(figsize=(16,6))
ax = sns.countplot(data=df , x='Occupation',palette=sns.color_palette("Set2"))
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
Product_Category_sales = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(20,6))
sns.barplot(data=Product_Category_sales , x='Product_Category',y='Amount',palette=sns.color_palette("Set2"))
plt.show()
```

## from above graph we can see that most of the sold product are from food, clothing and electronic category
